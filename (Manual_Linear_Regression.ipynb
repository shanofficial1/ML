```python
# Training data
x = [500, 1000, 1500]   # sizes (input)
y = [25, 50, 75]        # prices (output)

# Start values (random guess)
m = 0
c = 0

learning_rate = 0.0000001   # very small to prevent NaN

# Training loop
for _ in range(10000):
    dm = 0
    dc = 0

    for i in range(len(x)):
        y_pred = m * x[i] + c       # prediction
        error = y[i] - y_pred       # difference

        dm += -2 * x[i] * error
        dc += -2 * error

    # Average gradient and update
    m = m - learning_rate * (dm / len(x))
    c = c - learning_rate * (dc / len(x))

print("m =", m)
print("c =", c)


size = 2000
print("Predicted price:", m * size + c, "lakhs")
```
