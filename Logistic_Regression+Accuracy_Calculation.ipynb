```python
import math

# Training data
x = [1, 2, 3, 4, 5]
y = [0, 0, 1, 1, 1]

# From previous training
m = 1.5
c = -4     # (These are example values; use your trained values if you want)

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

def predict(x):
    return sigmoid(m * x + c)

def classify(p):
    return 1 if p >= 0.5 else 0

TP = TN = FP = FN = 0

for i in range(len(x)):
    prob = predict(x[i])
    y_pred = classify(prob)
    y_true = y[i]

    if y_true == 1 and y_pred == 1:
        TP += 1
    elif y_true == 0 and y_pred == 0:
        TN += 1
    elif y_true == 0 and y_pred == 1:
        FP += 1
    elif y_true == 1 and y_pred == 0:
        FN += 1

accuracy = (TP + TN) / (TP + TN + FP + FN)

print("TP:", TP)
print("TN:", TN)
print("FP:", FP)
print("FN:", FN)
print("Accuracy:", accuracy)
```
