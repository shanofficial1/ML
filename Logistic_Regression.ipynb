```python
import math

# Training data
x = [1, 2, 3, 4, 5]
y = [0, 0, 1, 1, 1]

m = 0   # slope
c = 0   # intercept
lr = 0.1   # learning rate

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

for epoch in range(2000):
    dm = 0
    dc = 0

    for i in range(len(x)):
        z = m * x[i] + c
        y_pred = sigmoid(z)
        error = y[i] - y_pred

        dm += -x[i] * error
        dc += -error

    m = m - lr * (dm / len(x))
    c = c - lr * (dc / len(x))

print("m =", m)
print("c =", c)

# Test Prediction
def predict(hours):
    return sigmoid(m * hours + c)

print("Prediction for 2.5 hours:", predict(2.5))
print("Prediction for 4 hours:", predict(4))

```
