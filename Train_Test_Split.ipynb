Train-Test Split

```python
x_train = [1, 2, 3, 4]
y_train = [0, 0, 1, 1]
```

```python
x_test = [5]
y_test = [1]
```

```python
import math

m = 0
c = 0
lr = 0.1  # learning rate

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

for epoch in range(2000):
    dm = 0
    dc = 0
    for i in range(len(x_train)):
        y_pred = sigmoid(m * x_train[i] + c)
        error = y_train[i] - y_pred
        dm += -x_train[i] * error
        dc += -error
    m -= lr * (dm / len(x_train))
    c -= lr * (dc / len(x_train))

print("Trained m:", m)
print("Trained c:", c)
import math

m = 0
c = 0
lr = 0.1  # learning rate

def sigmoid(z):
    return 1 / (1 + math.exp(-z))

for epoch in range(2000):
    dm = 0
    dc = 0
    for i in range(len(x_train)):
        y_pred = sigmoid(m * x_train[i] + c)
        error = y_train[i] - y_pred
        dm += -x_train[i] * error
        dc += -error
    m -= lr * (dm / len(x_train))
    c -= lr * (dc / len(x_train))

print("Trained m:", m)
print("Trained c:", c)
```

```python
def predict(x):
    return sigmoid(m * x + c)

prob = predict(x_test[0])
prediction = 1 if prob >= 0.5 else 0

print("\nTest Input:", x_test[0])
print("Prediction Probability:", prob)
print("Predicted Result:", prediction)
print("Actual Result:", y_test[0])
```
